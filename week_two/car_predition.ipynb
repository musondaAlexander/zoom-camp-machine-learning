```python
# make heads to lower case
df.columns = df.columns.str.lower()
# check the columns again
df.head(10)
```

```python
# re[lace spaces with underscores
df.columns = df.columns.str.replace(' ', '_')
# check the columns again
df.head(10) 
```

```python
# change all the string in all the columns to lower case
df.dtypes
```

```python
list =  df.columns
#get the string values instead of the objects
strings = df.select_dtypes(include=['object']).columns.tolist()
strings

list = strings
list 
```

```python
for i in list:
    df[i] = df[i].str.lower().str.replace(' ', '_') 
```
